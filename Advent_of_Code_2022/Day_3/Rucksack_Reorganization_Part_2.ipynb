```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")




maxlength=len(datafromfile)

#validate last item
print(datafromfile[maxlength-1])
#total records
print(f"total_numbers= {maxlength}")


#start programming

#initializing lists and counters
left = ["A"]*(maxlength)
right = ["A"]*(maxlength)
scoredict = {"a" : 1,"b" : 2,"c" : 3,"d" : 4,"e" : 5,"f" : 6,"g" : 7,"h" : 8,"i" : 9,"j" : 10,
             "k" : 11,"l" : 12,"m" : 13,"n" : 14,"o" : 15,"p" : 16,"q" : 17,"r" : 18,"s" : 19,"t" : 20,
             "u" : 21,"v" : 22,"w" : 23,"x" : 24,"y" : 25,"z" : 26,"A" : 27,"B" : 28,"C" : 29,"D" : 30,
             "E" : 31,"F" : 32,"G" : 33,"H" : 34,"I" : 35,"J" : 36,"K" : 37,"L" : 38,"M" : 39,"N" : 40,
             "O" : 41,"P" : 42,"Q" : 43,"R" : 44,"S" : 45,"T" : 46,"U" : 47,"V" : 48,"W" : 49,"X" : 50,
             "Y" : 51,"Z" : 52}
print(scoredict.get("b"))
score=0
flag = 0
n=0 #which item we're talking about
x=0 #dimension 1, iterate through first item
y=0 #dimension 2, check against second item
z=0 #dimension 3, if second one succeeds, check on last item
#for every 3 items, look through others
while n < maxlength:
    flag=0
    x=0
    y=0
    z=0
    print(f"n value is {n}")
    while x < len(datafromfile[n]):
        y=0
        z=0
        if flag == 1:    #if flag is found
            x = len(datafromfile[n])    #set x to max to escape loop
        print(f"layer 1: check {x} against {len(datafromfile[n])}")

        while y < len(datafromfile[n+1]):
            z=0
            print(f"checking {datafromfile[n][x-1]} == {datafromfile[n+1][y-1]}")
            if flag == 1:
                y = len(datafromfile[n+1])
            elif datafromfile[n][x] == datafromfile[n+1][y]: #condition is found for 1st and second. check 2nd against 3rd
                while z < len(datafromfile[n+2]):
                    if datafromfile[n+1][y] == datafromfile[n+2][z]:
                        flag = 1    #condition found! raise the flag.
                        print(f"Flag is found! {datafromfile[n+1][y]}")
                        score = score + scoredict.get(datafromfile[n+1][y])
                        print(f"new score is : {score}")
                        z = len(datafromfile[n+2])    #escape the loop
                    else:
                        z=z+1
                
            y = y+1
        
        x = x+1
    n=n+3
    




                
print(score)
            
```
