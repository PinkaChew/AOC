```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")




maxlength=len(datafromfile)

#validate last item
print(datafromfile[maxlength-1])
#total records
print(f"total_numbers= {maxlength}")


#start programming

#initializing lists and counters
left = ["A"]*(maxlength)
right = ["A"]*(maxlength)
scoredict = {"a" : 1,"b" : 2,"c" : 3,"d" : 4,"e" : 5,"f" : 6,"g" : 7,"h" : 8,"i" : 9,"j" : 10,
             "k" : 11,"l" : 12,"m" : 13,"n" : 14,"o" : 15,"p" : 16,"q" : 17,"r" : 18,"s" : 19,"t" : 20,
             "u" : 21,"v" : 22,"w" : 23,"x" : 24,"y" : 25,"z" : 26,"A" : 27,"B" : 28,"C" : 29,"D" : 30,
             "E" : 31,"F" : 32,"G" : 33,"H" : 34,"I" : 35,"J" : 36,"K" : 37,"L" : 38,"M" : 39,"N" : 40,
             "O" : 41,"P" : 42,"Q" : 43,"R" : 44,"S" : 45,"T" : 46,"U" : 47,"V" : 48,"W" : 49,"X" : 50,
             "Y" : 51,"Z" : 52}
x=0
print(scoredict.get("V"))


#populating left list and right list
while x < maxlength:
    strlen = int(len(datafromfile[x])/2-1)
    left[x] = datafromfile[x][0:strlen]
    right[x] = datafromfile[x][strlen+1:len(datafromfile[x])]
    print(f"{datafromfile[x]}")
    print(f"{left[x]}")
    print(f"{right[x]}")
    x=x+1
    
#check if left = right


print(f"left = {left}")
print(f"right = {right}")

sum_priority = 0
x=0
y=0
z=0


#for every left item, interate through left list
while x < maxlength:
    print(f"checking {datafromfile[x]}")
    y=0
    z=0
    print(f"dropping flag")
    print(f"{left[x]}")
    print(f"{right[x]}")
    #flag checks for if duplicated
    flag=0
    while y<len(left[x]):
        z=0
        print(f"check {left[x][y]}")
        while z<len(right[x]):
            if left[x][y] == right[x][z] and flag == 0:
                print(f"against {right[x][z]}")
                sum_priority = sum_priority + scoredict.get(right[x][z])
                print(f"found it! raising flag and adding {scoredict.get(right[x][z])}")
                flag = 1
                print(sum_priority)
            z=z+1
        y = y+1
    x=x+1



                
print(sum_priority)
            
```
