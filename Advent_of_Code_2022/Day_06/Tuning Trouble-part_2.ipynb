```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = str(rawdata.split("\n")).replace("[","").replace("]","").replace("'","")




maxlength=len(datafromfile)

#validate last item
print(datafromfile[maxlength-1])
#total records
print(f"total_numbers= {maxlength}")
print(f"A: {datafromfile[0]}")


def check_unique(str):
    for i in range(len(str)):
        for j in range(i + 1,len(str)):
            if(str[i] == str[j]):
                return False
    return True

#start programming
n = 0 #Counter to iterate through string
while n<maxlength:
    #print(f"start : {datafromfile[n]}")
    #print(f"end : {datafromfile[n+3]}")    
    string_check = str(datafromfile[n:n+14])
    print(string_check)
    if check_unique(string_check):
        break
    else:
        n=n+1
print(f"final result = {datafromfile[n]}{datafromfile[n+1]}{datafromfile[n+2]}{datafromfile[n+3]}")
print(f"number of characters = {n+14}")
```
