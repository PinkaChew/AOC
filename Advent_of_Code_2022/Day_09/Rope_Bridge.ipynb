```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")




maxlength=len(datafromfile)

#print(f"rawfile = {datafromfile}")

#validate last item
#print(datafromfile[maxlength-1])
#total records
#print(f"total_numbers= {maxlength}")
#print(f"len row: {len(datafromfile[0])}")
#print(f"len height: {len(datafromfile)}")

#populating commands
commandlist = []
i=0
while i < len(datafromfile):
    templist=[]
    templist.append(datafromfile[i][0])
    templist.append(datafromfile[i][2:])
    commandlist.append(templist)
    i=i+1

print(f"commandlist is now {commandlist}")





###########################

def move_locations(d):
    #move head
    global locations
    if d == "U":
        locations["H"][0] -= 1
        refresh_tail()
    elif d =="D":
        locations["H"][0] += 1
        refresh_tail()
    elif d =="L":
        locations["H"][1] -= 1
        refresh_tail()
    elif d =="R":
        locations["H"][1] += 1
        refresh_tail()
    print(f"new location is now {locations}")

###########################

def refresh_tail():
    global locations
    global t_visited
    xdiff = locations["H"][1] - locations["T"][1]
    ydiff = locations["H"][0] - locations["T"][0]
    #start check if H vs T is 2 spaces apart horizontally or vertically
    if xdiff >= 2 and ydiff == 0:
        locations["T"][1] += 1
    elif xdiff <= (-2) and ydiff == 0:
        locations["T"][1] -= 1
    elif ydiff >= 2 and xdiff == 0:
        locations["T"][0] += 1
    elif ydiff <= (-2) and xdiff == 0:
        locations["T"][0] -= 1
        
    #check if H vs T is 2 spaces apart but need to move diagonally
    elif xdiff >= 2 and ydiff == 1:
        locations["T"][1] += 1
        locations["T"][0] += 1
    elif xdiff >= 2 and ydiff == -1:
        locations["T"][1] += 1
        locations["T"][0] -= 1
    elif ydiff >= 2 and xdiff == 1:
        locations["T"][0] += 1
        locations["T"][1] += 1
    elif ydiff >= 2 and xdiff == -1:
        locations["T"][0] += 1
        locations["T"][1] -= 1
    elif xdiff <= (-2) and ydiff == 1:
        locations["T"][1] -= 1
        locations["T"][0] += 1
    elif xdiff <= (-2) and ydiff == -1:
        locations["T"][1] -= 1
        locations["T"][0] -= 1
    elif ydiff <= (-2) and xdiff == 1: 
        locations["T"][0] -= 1
        locations["T"][1] += 1
    elif ydiff <= (-2) and xdiff == -1: 
        locations["T"][0] -= 1
        locations["T"][1] -= 1
    #end of movements!
    t_visited.add(str(locations.get("T")[0]) + "x" + str(locations.get("T")[1]))
        
###########################


locations = {
    "H": [0,0],
    "T": [0,0]
            }
t_visited = {0x0}
print(f"t_visited is now {t_visited}")


x=0
while x < len(commandlist):
    d = commandlist[x][0]
    counter = int(commandlist[x][1])
    y=0
    while y<counter:
        move_locations(d)
        y += 1
    x += 1

print(f"t_visited is now {len(t_visited)-1}")
```
