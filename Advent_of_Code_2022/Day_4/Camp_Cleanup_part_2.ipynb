```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")




maxlength=len(datafromfile)

#validate last item
print(datafromfile[maxlength-1])
#total records
print(f"total_numbers= {maxlength}")
print(f"A: {datafromfile[0]}")
print(f"B: {datafromfile[1]}")


#start programming
listA = ["A"]*maxlength
listB = ["A"]*maxlength

x=0
#split items up into two lists
while x < maxlength:
    listA[x] = datafromfile[x].split(",")[0].split("-")
    listB[x] = datafromfile[x].split(",")[1].split("-")
    #print(f"listA: {listA[x]}")
    #print(f"listB: {listB[x]}")
    x=x+1
    
#check for overlaps
x=0
overlapcounter = 0
while x < maxlength:
    #found condition, B approaches A from left
    if int(listB[x][1]) >= int(listA[x][0]) >= int(listB[x][0]):    
        overlapcounter = overlapcounter + 1
    #found condition, B subset of A
    elif int(listA[x][1]) <= int(listB[x][0]) <= int(listB[x][1]) <= int(listA[x][0]):    
        overlapcounter = overlapcounter + 1
    #found condition, B approaches A from right
    elif int(listB[x][1]) <= int(listA[x][1]) <= int(listB[x][0]):
        overlapcounter = overlapcounter + 1
    #UNO-REVERSE!
    #found condition, A approaches B from left
    elif int(listA[x][1]) >= int(listB[x][0]) >= int(listA[x][0]):    
        overlapcounter = overlapcounter + 1
    #found condition, A subset of B
    elif int(listB[x][1]) <= int(listA[x][0]) <= int(listA[x][1]) <= int(listB[x][0]):     
        overlapcounter = overlapcounter + 1
    #found condition, A approaches B from right
    elif int(listA[x][1]) <= int(listB[x][1]) <= int(listA[x][0]):
        overlapcounter = overlapcounter + 1
    else:
        pass
    x = x+1
    
print(f"Total supersets = {overlapcounter}")


```
