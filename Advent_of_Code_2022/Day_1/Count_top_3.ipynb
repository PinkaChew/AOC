```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.splitlines(True)


maxlength=len(datafromfile)

print(datafromfile[maxlength-1])

print(f"total_numbers= {maxlength}")

largest_number = 0
second_largest = 0
third_largest = 0
x=0
y=0
sum_array = list(range(0,maxlength))
while x < maxlength:
    if datafromfile[x] == "\n":
        if sum_array[y]>largest_number:
            third_largest = second_largest
            second_largest = largest_number
            largest_number = sum_array[y]
        elif sum_array[y]>second_largest:
            third_largest = second_largest
            second_largest = sum_array[y]
        elif sum_array[y]>third_largest:
            third_largest = sum_array[y]
        else:
            pass
        y=y+1
        x=x+1
        sum_array[y]=0
        print(largest_number)
    else:
        sum_array[y] = sum_array[y]+ int(datafromfile[x])
        x=x+1

top_three = third_largest + second_largest + largest_number
print(f"the sum of top three number is {top_three}")
    
```
