```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")

x=0
tvscreen = []
while x < 7:
    templist = [""]*40
    tvscreen.append(templist)
    x+=1

row = 0
counter = 1
signal_sum = 0
rowcount = 1


def restartrow():
    global counter
    if counter == 41 or counter == 81 or counter == 121 or counter == 161 or counter == 201 or counter == 241:
        global rowcount
        global row
        rowcount = 1
        row += 1
    else:
        return

def addx(v):
    global counter
    global x
    draw_pix()
    counter += 1
    restartrow()
    draw_pix()
    counter += 1
    restartrow()
    x += v
    
    
def noop():
    global counter
    draw_pix()
    counter += 1
    restartrow()
    
def draw_pix():
    global rowcount
    global row
    global x
    print(f"x is now {x}")
    print(f"row is now {row}")
    print(f"rowcount is now {rowcount}")

    if rowcount - x >= 0 and rowcount - x <= 2:
        tvscreen[row][rowcount-1] = "#"
        rowcount += 1
    else:
        tvscreen[row][rowcount-1] = "."
        rowcount += 1
    
    
z=0
x=1
while z<len(datafromfile):
    if datafromfile[z][0:4] == "addx":
        addx(int(datafromfile[z][5:]))
    elif datafromfile[z][0:4] == "noop":
        noop()
    z += 1
print(tvscreen[0])
print(tvscreen[1])
print(tvscreen[2])
print(tvscreen[3])
print(tvscreen[4])
print(tvscreen[5])
```
