```python
#importing the os module
import os
import queue as q
import math

#to get the current working directory
directory = os.getcwd()

print(directory)

filepath = directory+"/input.txt"
print(filepath)

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")

#print(datafromfile)

#print(f"the length is {len(datafromfile)}")





################################################################################################
def createitemqueue(x):
    global datafromfile
    global monkeydict
    itemqueue = q.Queue()
    itemstring = datafromfile[x*7 + 1][17:].replace(" ","").split(",")
    n=0
    while n < len(itemstring):
        itemqueue.put(int(itemstring[n]))
        n+=1
    return itemqueue
################################################################################################

def defineoperation(x):
    global datafromfile
    last_space = 0
    n=0
    while n < len(datafromfile[x*7 + 2]):
        if datafromfile[x*7 + 2][n] == "=":
            last_space = n
        n += 1
    operationval = (datafromfile[x*7 + 2][last_space+2:])
    return operationval
################################################################################################
def definetest(x):
    global datafromfile
    last_space = 0
    n=0
    while n < len(datafromfile[x*7 + 3]):
        if datafromfile[x*7 + 3][n] == " ":
            last_space = n
        n += 1
    testval = int((datafromfile[x*7 + 3][last_space+1:]))
    return testval
################################################################################################
def definetrue(x):
    global datafromfile
    last_space = 0
    n=0
    while n < len(datafromfile[x*7 + 4]):
        if datafromfile[x*7 + 4][n] == " ":
            last_space = n
        n += 1
    trueval = int((datafromfile[x*7 + 4][last_space+1:]))
    return trueval
################################################################################################
def definefalse(x):
    global datafromfile
    last_space = 0
    n=0
    while n < len(datafromfile[x*7 + 5]):
        if datafromfile[x*7 + 5][n] == " ":
            last_space = n
        n += 1
    falseval = int((datafromfile[x*7 + 5][last_space+1:]))
    return falseval
################################################################################################
def create_monkey(x):
    tempdict = [createitemqueue(x),defineoperation(x),definetest(x),definetrue(x),definefalse(x),0]
    #monkeydict.add(int(x),tempdict)
    print(f"tempdict is {tempdict}")
    return tempdict
    
    
################################################################################################
def monkey_inspections(x):
    global monkeydict
    global splitat
    global leftop
    global rightop
    global tempvalue
    global n
    monkeydict[x][5] += 1
    tempvalue = int(monkeydict[x][0].get())
    n=0
    while n < len(monkeydict[x][1]):
        if monkeydict[x][1][n] == "+" or monkeydict[x][1][n] == "*":
            operator = monkeydict[x][1][n]
            splitat = n
        n += 1
    leftop = monkeydict[x][1][:splitat-1]
    rightop = monkeydict[x][1][splitat+2:]
    if leftop == "old":
        leftop = int(tempvalue)
    else: 
        leftop = int(leftop)
        
    if rightop == "old":
        rightop = int(tempvalue)
    else: 
        rightop = int(rightop)
        
    if operator == "*":
        tempvalue = leftop*rightop
    elif operator == "+":
        tempvalue = leftop+rightop
        
#    tempvalue = math.floor(tempvalue / 3)
    
    if tempvalue%int(monkeydict[x][2]) == 0:
        monkeydict[int(monkeydict[x][3])][0].put(tempvalue)
    else:
        monkeydict[int(monkeydict[x][4])][0].put(tempvalue)
    
    return
################################################################################################

monkeycount = (len(datafromfile)-1)/7
monkeydict = {}
counter = 0 
monkeys = 0
while monkeys < monkeycount:
    monkeydict[monkeys] =  create_monkey(monkeys)
    monkeys += 1

    

    
monkeycounter = 0
rounds = 0
while rounds < 100:
    x=0
    print(f"rounds is {rounds}")
    while x < monkeycount:
        while monkeydict[x][0].qsize() > 0:
            print(f"qsize is {monkeydict[x][0].qsize()}")
            monkey_inspections(x)
        x+=1
    
    rounds += 1


x=0
while x < monkeycount:
    print(f"{x} monkey has {monkeydict[x][5]}")
    x+=1
print(f"{336*337}")

#print(f"monkeys0 divisible by {monkeys.get(0)[1]}")
#print(f"monkeys0 old_queue {monkeys.get(0)[0]}")
#thrown = 69
#monkeys[0][0] = queue.Queue(thrown)
#print(f"monkeys0 new_queue {monkeys.get(0)[0]}")

#lista = q.Queue()
#print(lista)
#lista.put(69)
#print(lista.get())
#print(lista)
```
