```python
#importing the os module
import os

#to get the current working directory
directory = os.getcwd()


filepath = directory+"/input.txt"

#filepath is just the path + name
thefile=open(filepath)
#throw the raw format into variable rawdata
rawdata=thefile.read()
#split by "\n", which stands for new line
datafromfile = rawdata.split("\n")




maxlength=len(datafromfile)


#validate last item
#total records

forest = []

#planting the trees

for row in datafromfile:
    templist = []
    for char in row:
        templist.append(char)
    forest.append(templist)

################
def check_taller_left(x,y):
    global forest
    i=y-1
    while i>=0:
        if forest[x][i] >= forest[x][y]:
            return True
        else:
            i=i-1
    return False

def check_taller_right(x,y):
    global forest
    i=y+1
    while i<len(forest[x]):
        if forest[x][i] >= forest[x][y]:
            return True
        i=i+1
    return False

def check_taller_up(x,y):
    global forest
    i=x-1
    while i>=0:
        if forest[i][y] >= forest[x][y]:
            return True
        else:
            i=i-1
    return False

def check_taller_down(x,y):
    global forest
    i=x+1
    while i<len(forest):
        if forest[i][y] >= forest[x][y]:
            return True
        i=i+1
    return False

def check_hidden(x,y):
    if check_taller_up(x,y) and check_taller_down(x,y) and check_taller_left(x,y) and check_taller_right(x,y):
        return True
    else:
        return False
###########################

x=0
y=0
counter = 0
while x<len(forest):
    y=0
    while y<len(forest):
        if check_hidden(x,y):
            pass
        else:
            counter = counter+1
        y=y+1
    x=x+1

print(f"Visible trees = {counter}")
```
